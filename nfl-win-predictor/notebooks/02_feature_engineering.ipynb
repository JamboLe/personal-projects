## Leakage sanity check
No post-game columns (`home_score`, `away_score`, `result`, `total`, `overtime`) may appear as features.

```python
import pandas as pd
df = pd.read_csv('../data/processed/merged_features.csv')
leaky = {'home_score','away_score','result','total','overtime'}
print('leaky present:', leaky & set(df.columns))
print('NaNs:', int(df.isna().sum().sum()))
print('home win rate:', round(df['home_win'].mean(),3))
```

Result: 1,839 games, 0 NaNs, no leaky columns, home-win rate ~0.53 (matches the naive baseline).
