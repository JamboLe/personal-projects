### Findings to talk about in the interview
1. **Every model beats the 0.535 baseline** — our engineered features carry real signal.
2. **Logistic Regression won.** On a ~1,300-row, mostly-linear tabular problem, the simplest model generalized best; the tree ensembles slightly overfit. This is exactly why you always train a simple baseline — fancier isn't automatically better.
3. **Adding the spread lifted accuracy ~4 points** (0.636 -> 0.675), pulling us toward the ~0.66 market line. That gain is the model absorbing Vegas's prediction, not new skill of ours — which is why we keep the nospread model as the honest result.

> Note: the app loads XGBoost nospread by default (per the spec) because it gives well-calibrated probabilities and SHAP explanations. The LogReg win is reported honestly as the best raw accuracy.
