# 04 — SHAP Interpretability

SHAP explains *why* the model made each prediction by attributing the outcome to individual features. We use `TreeExplainer` on the XGBoost model.

```python
import json, pickle
import pandas as pd, shap
model = pickle.load(open('../models/xgboost_nospread.pkl','rb'))
features = json.load(open('../models/feature_list_nospread.json'))
test = pd.read_csv('../data/processed/test.csv')
X = test[features]
explainer = shap.TreeExplainer(model)
sv = explainer.shap_values(X)
shap.summary_plot(sv, X)
```

## Mean |SHAP| feature ranking (test set)

| Rank | Feature | Mean |SHAP| |
|---|---|---|
| 1 | points_allowed_roll5_diff (defense edge) | 0.276 |
| 2 | rush_yards_roll5_diff | 0.191 |
| 3 | pass_yards_roll5_diff | 0.163 |
| 4 | turnovers_roll5_diff | 0.105 |
| 5 | temp | 0.097 |

**Takeaway:** the model leans hardest on *recent scoring defense*, then the offensive yardage edges and turnover margin — all features WE engineered. That's the evidence our feature work, not luck, drives the predictions.

Per-game SHAP values power the 'What drove this prediction' list in the app.
